# Lists


```python
list1 = [1, 0]
list2 = [1, 0]

print(list1 == list2) # The values are the same -> True
print(list1 is list2) # They are two different objects in memory -> False
print(list1 is not list2) # Reverse of the last one
```

***Notice***: Lists are oriented

```python
a = [1, 2, 3]
b = [3, 2, 1]
c = [1, 3, 2]

print(a == b)
print(a == c)
print(c == b)
```

# `is` Operator

```python
a = 1
b = 1
print(a is b)

x = 257
y = 257
print(x is y)
```

***Notice***: By default numbers from -5 to 256 are reserved and cached in memory. So all numbers in this range are the same object in memory
