## Variables Naming Rules

##### There are several specific rules that programmers should follow:

1. A variable name should start with either alphabetic characters (a-z or A-Z) or underscores (_)
2. A variable name can only contain alphabetic characters (a-z or A-Z) or numerical digits (0-9) or underscores (_)
3. A variable name shouldn't be the same as a reserved python keyword (e.g. if, for, while, print and etc.)

*Notice*: Variable names are case sensitive. Therefore `a` is considered different than `A` and they are two separate variables

## Operators in Python

### Comparison Operators

* Is a equal to b? (a == b)

* Is a not equal to b? (a != b)

* Is a *strictly* greater than b? (a > b)

* Is a greater than or equal to b? (a >= b)

* Is a *strictly* less than b? (a < b)

* Is a less than or equal to b? (a <= b)

```python
a = (2==2)
b = (5/2 > 2.3)
c = (2**9 > 3**8)

print(a)
print(b)
print(c)
```

### Arithmetic Operators

* Addition -> **x + y**
* Subtraction -> **x - y**
* Multiplication -> **x * y**
* Division -> **x / y**
* Modules (remainder of division) -> **x % y**
* Exponentiation (x to the power of y) -> **x ** y**
* Floor division (Integer division) -> **x // y**

Other arithmetic operators (such as //, **, %) can be combined by equal sign, similarly

### Logical Operators

* `a and b` - Returns `True` if both statements `a` and `b` are `True`
* `a or b` - Returns `True` if at least one of the statements `a` or `b` are `True`
* `not a` - Reverses the boolean value of `a`
