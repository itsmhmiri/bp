# راهنمای جامع Markdown و HTML

این راهنما یک مرور کامل از دستورات رایج Markdown و تگ‌های مفید HTML را ارائه می‌دهد که می‌توانند در Markdown برای قالب‌بندی پیشرفته‌تر استفاده شوند.

## ۱. سرتیترها

سرتیترها برای ساختاردهی به سند شما استفاده می‌شوند. از `#` برای H1، `##` برای H2 و به همین ترتیب استفاده کنید.

```markdown
# این یک سرتیتر سطح ۱ است
## این یک سرتیتر سطح ۲ است
### این یک سرتیتر سطح ۳ است
#### این یک سرتیتر سطح ۴ است
##### این یک سرتیتر سطح ۵ است
###### این یک سرتیتر سطح ۶ است
```

### نتیجه:
# این یک سرتیتر سطح ۱ است
## این یک سرتیتر سطح ۲ است
### این یک سرتیتر سطح ۳ است
#### این یک سرتیتر سطح ۴ است
##### این یک سرتیتر سطح ۵ است
###### این یک سرتیتر سطح ۶ است

## ۲. تأکید

می‌توانید با استفاده از پررنگ، مورب و خط‌خورده به متن خود تأکید بدهید.

| سبک          | دستور               | مثال                    |
|--------------|---------------------|-------------------------|
| پررنگ        | `**متن**` یا `__متن__` | **این پررنگ است**      |
| مورب         | `*متن*` یا `_متن_`     | *این مورب است*         |
| خط‌خورده     | `~~متن~~`           | ~~این خط‌خورده است~~   |
| پررنگ و مورب | `***متن***`         | ***این پررنگ و مورب است*** |

## ۳. فهرست‌ها

### فهرست‌های بدون شماره
از `*`، `-` یا `+` برای ایجاد فهرست‌های بدون شماره استفاده کنید.

```markdown
* آیتم ۱
* آیتم ۲
  * آیتم تو در تو ۲الف
  * آیتم تو در تو ۲ب
```

### نتیجه:
* آیتم ۱
* آیتم ۲
  * آیتم تو در تو ۲الف
  * آیتم تو در تو ۲ب

### فهرست‌های شماره‌دار
از اعداد به دنبال یک نقطه استفاده کنید.

```markdown
1. آیتم اول
2. آیتم دوم
3. آیتم سوم
```

### نتیجه:
1. آیتم اول
2. آیتم دوم
3. آیتم سوم

## ۴. لینک‌ها

با قرار دادن متن لینک در کروشه `[]` و آدرس URL در پرانتز `()` یک لینک ایجاد کنید.

```markdown
[بازدید از گوگل](https://www.google.com)
```

### نتیجه:
[بازدید از گوگل](https://www.google.com)

## ۵. تصاویر

دستور تصاویر شبیه لینک‌ها است، اما با یک علامت تعجب `!` در ابتدا.

```markdown
![لوگوی مارک‌داون](https://markdown-here.com/img/icon256.png)
```

### نتیجه:
![لوگوی مارک‌داون](https://markdown-here.com/img/icon256.png)

## ۶. کد

### کد درون‌خطی
کد را با بک‌تیک تکی `` ` `` احاطه کنید.

`console.log('سلام دنیا!');`

### بلوک‌های کد
کد خود را با سه بک‌تیک `` ``` `` احاطه کنید. همچنین می‌توانید زبان برنامه‌نویسی را برای برجسته‌سازی نحو مشخص کنید.

```javascript
function greet(name) {
  return "سلام، " + name + "!";
}
```

### نتیجه:
```javascript
function greet(name) {
  return "سلام، " + name + "!";
}
```

## ۷. نقل‌قول‌ها

از کاراکتر `>` برای ایجاد نقل‌قول استفاده کنید.

```markdown
> این یک نقل‌قول است. می‌توانید چندین پاراگراف داشته باشید.
>
> مثل این.
```

### نتیجه:
> این یک نقل‌قول است. می‌توانید چندین پاراگراف داشته باشید.
>
> مثل این.

## ۸. خطوط افقی

با استفاده از سه یا بیشتر خط تیره `---`، ستاره `***` یا زیرخط `___` یک خط افقی ایجاد کنید.

```markdown
---
***
___
```

### نتیجه:
---

## ۹. جداول

با استفاده از خط عمودی `|` و خط تیره `-` جدول ایجاد کنید.

```markdown
| سرستون ۱ | سرستون ۲ | سرستون ۳ |
|:---------|:--------:|---------:|
| چپ‌چین   | وسط‌چین  | راست‌چین |
| سلول ۲   | سلول ۳   | سلول ۴   |
```

### نتیجه:
| سرستون ۱ | سرستون ۲ | سرستون ۳ |
|:---------|:--------:|---------:|
| چپ‌چین   | وسط‌چین  | راست‌چین |
| سلول ۲   | سلول ۳   | سلول ۴   |

## ۱۰. تگ‌های مفید HTML

می‌توانید از HTML خام در Markdown خود برای قالب‌بندی پیشرفته‌تر استفاده کنید.

### تگ‌های `<details>` و `<summary>`
بخش‌های قابل جمع‌شونده ایجاد کنید.
<details>
  <summary>برای باز کردن کلیک کنید!</summary>
  
  این محتوا تا زمانی که روی خلاصه کلیک نکنید پنهان است.
</details>

### تگ `<img>` با ویژگی‌های اندازه
اندازه یک تصویر را کنترل کنید.
<img src='https://markdown-here.com/img/icon256.png' width='100' height='100'>

### تگ `<br>`
شکست خط اجباری.
این خط اول است.<br>این خط دوم است.

### تگ `<font>`
رنگ متن را تغییر دهید (با احتیاط استفاده کنید).
<font color='red'>این متن قرمز است.</font><br>
<font color='blue'>این متن آبی است.</font>

### تگ‌های `<u>` و `<mark>`
متن را زیرخط‌دار یا برجسته کنید.
<u>این متن زیرخط‌دار است.</u><br>
<mark>این متن برجسته است.</mark>
